```python
import pandas as pd

def read_csv(file: str):
    return pd.read_csv(file, header=0)
```

```python
network = read_csv(r"../network.csv")
network
```

```python
to_language_count = network.groupby("to_language").count().sort_values("sponsor", ascending=False)
to_language_count
```

```python
to_language_count.plot.bar(fontsize=10, width=0.6, figsize=(12, 6), y="to", xlabel="lang", ylabel="sponsored count")
```
